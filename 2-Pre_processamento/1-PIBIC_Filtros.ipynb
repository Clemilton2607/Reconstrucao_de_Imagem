# Aplicando Filtro nos datasets


## 1- Filtro Vermelho (620-750 nm)

```python
import os
from PIL import Image

def aplicar_filtro_vermelho_diretorio(diretorio_entrada, diretorio_saida):
    # Cria o diretório de saída, se ainda não existir
    if not os.path.exists(diretorio_saida):
        os.makedirs(diretorio_saida)
    
    # Lista todos os arquivos no diretório de entrada
    arquivos = os.listdir(diretorio_entrada)
    
    # Itera sobre cada arquivo
    for arquivo in arquivos:
        # Verifica se é um arquivo de imagem
        if arquivo.endswith('.png') or arquivo.endswith('.jpg') or arquivo.endswith('.jpeg'):
            # Caminho completo do arquivo de entrada
            caminho_arquivo_entrada = os.path.join(diretorio_entrada, arquivo)
            
            try:
                # Aplica o filtro vermelho na imagem
                img = aplicar_filtro_vermelho(caminho_arquivo_entrada)
                
                # Caminho completo do arquivo de saída
                nome_arquivo_saida = os.path.splitext(arquivo)[0] + '_filtrada.png'
                caminho_arquivo_saida = os.path.join(diretorio_saida, nome_arquivo_saida)
                
                # Salva a imagem filtrada no diretório de saída
                img.save(caminho_arquivo_saida)
                
                print(f"Filtro vermelho aplicado com sucesso em {arquivo}!")
            except Exception as e:
                print(f"Erro ao aplicar o filtro em {arquivo}: {str(e)}")

def aplicar_filtro_vermelho(imagem):
    # Abre a imagem
    img = Image.open(imagem)
    
    # Obtém as dimensões da imagem
    largura, altura = img.size
    
    # Cria uma nova imagem com o mesmo tamanho da original
    nova_imagem = Image.new('RGB', (largura, altura))
    
    # Percorre cada pixel da imagem original
    for x in range(largura):
        for y in range(altura):
            # Obtém o valor RGB do pixel
            r, g, b = img.getpixel((x, y))
            
            # Define a cor vermelha como o valor máximo de R e 0 para G e B
            nova_cor = (r, 0, 0)
            
            # Define o novo valor do pixel na nova imagem
            nova_imagem.putpixel((x, y), nova_cor)
    
    return nova_imagem

# Exemplo de uso
diretorio_entrada = r'D:\Dataset_PIBIC\Coloridas'
diretorio_saida = r'D:\Dataset_PIBIC\Vermelho'
aplicar_filtro_vermelho_diretorio(diretorio_entrada, diretorio_saida)
```

## 2- Filtro Laranja (590-620 nm)

```python
import os
from PIL import Image

def aplicar_filtro_laranja_diretorio(diretorio_origem, diretorio_destino):
    # Verifica se o diretório de destino existe, caso contrário, cria-o
    if not os.path.exists(diretorio_destino):
        os.makedirs(diretorio_destino)

    # Lista todos os arquivos no diretório de origem
    arquivos = os.listdir(diretorio_origem)

    # Itera sobre cada arquivo
    for arquivo in arquivos:
        # Verifica se é um arquivo de imagem
        if arquivo.endswith('.png') or arquivo.endswith('.jpg') or arquivo.endswith('.jpeg'):
            # Caminho completo do arquivo de origem
            caminho_arquivo_origem = os.path.join(diretorio_origem, arquivo)

            try:
                # Aplica o filtro laranja na imagem
                img = aplicar_filtro_laranja(caminho_arquivo_origem)

                # Caminho completo do arquivo de destino
                nome_arquivo_destino = os.path.splitext(arquivo)[0] + '_filtrada.png'
                caminho_arquivo_destino = os.path.join(diretorio_destino, nome_arquivo_destino)

                # Salva a imagem filtrada no diretório de destino
                img.save(caminho_arquivo_destino)

                print(f"Filtro laranja aplicado com sucesso em {arquivo}!")
            except Exception as e:
                print(f"Erro ao aplicar o filtro em {arquivo}: {str(e)}")

def aplicar_filtro_laranja(imagem):
    # Abre a imagem
    img = Image.open(imagem)

    # Obtém as dimensões da imagem
    largura, altura = img.size

    # Cria uma nova imagem com o mesmo tamanho da original
    nova_imagem = Image.new('RGB', (largura, altura))

    # Percorre cada pixel da imagem original
    for x in range(largura):
        for y in range(altura):
            # Obtém o valor RGB do pixel
            r, g, b = img.getpixel((x, y))

            # Define a cor laranja com base nos valores RGB originais
            # A cor laranja é uma mistura de vermelho e verde, então definimos R como seu valor original, G como metade de seu valor original e B como 0
            nova_cor = (r, int(g / 2), 0)

            # Define o novo valor do pixel na nova imagem
            nova_imagem.putpixel((x, y), nova_cor)

    return nova_imagem

# Exemplo de uso
diretorio_origem = r'D:\Dataset_PIBIC\Coloridas'
diretorio_destino = r'D:\Dataset_PIBIC\Laranja'
aplicar_filtro_laranja_diretorio(diretorio_origem, diretorio_destino)
```
